```python
import json

with open("ia_generative.json", "r", encoding="utf-8") as f:
    json_data = json.load(f)

print(json_data)
```

```python
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter

json_splitter = RecursiveJsonSplitter(max_chunk_size=100)
json_chunks = json_splitter.split_json(json_data)
```

```python
json_chunks
```

```python
for chunk in json_chunks:
    print(chunk)
```

```python
# Splitter in output document

docs = json_splitter.create_documents(texts=json_chunks)
```

```python
for doc in docs:
    print(doc)
```
