```python
from langchain_openai import OpenAIEmbeddings

embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "This is a tutorial on OpenAI Embedding"

query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

Ce vecteur d'une longueur de 3072 représente le sens global de la phrase

```python
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)

text = "This is a tutorial on OpenAI Embedding"

query_result=embeddings_1024.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
# Text loader

from langchain_community.document_loaders import TextLoader

loader=TextLoader('speech.txt')

docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50) # L'overlap permet d'avoir un nombre de derniers caractères au début du groupe d'après
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
# Vector Embedding  and vector storeDB

from langchain_community.vectorstores import Chroma

# embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)

db = Chroma.from_documents(final_documents, embeddings_1024)
```

```python
db
```

```python
# Retrieve the results from query vecrorstoredb
query = "It will be all the easier for us to conduct ourselves as belligerents"

retrieved_result = db.similarity_search(query)
print(retrieved_result)
```
