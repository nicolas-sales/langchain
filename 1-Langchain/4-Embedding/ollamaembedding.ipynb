```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model='gemma:2b')  # By default : llama2
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first letter of Greek alphabet"
        "Beta is the second letter of Greak alphabet"
    ]
)
```

```python
r1
```

```python
len(r1[0])
```

```python
embeddings.embed_query('What is the second letter of Greak alphabet?')
```
